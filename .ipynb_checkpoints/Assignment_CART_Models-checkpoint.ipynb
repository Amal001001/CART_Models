## CART Model

```python
# Create an new dataframe with initialized values 

numrows = 200

decision_tree_df = pd.DataFrame({'height':[0.0]*numrows,
                             'weight':[0.0]*numrows,
                             'animal_type':[0.0]*numrows})

decision_tree_df.head()
```

```python
# Sort columns for convenience 
seq = ['height',
       'weight',
       'animal_type']

# Create a simple 3 feature dataset to use for demo
decision_tree_df = decision_tree_df.reindex(columns=seq)

decision_tree_df.head()
```

```python
# copy values from dog_horse dataset to decision_tree_df

# Note: we are copying the height and weight and target values from dog_horse dataset

for i in range(0 , numrows):
    height = df.iloc[i,0]
    decision_tree_df.iat[i,0] = height
    
    weight = df.iloc[i,1]
    decision_tree_df.iat[i,1] = weight
    
    animal_type = df.iloc[i,-2]
    decision_tree_df.iat[i,2] = animal_type

decision_tree_df = decision_tree_df.dropna() #exclude any rows with missing data
```

```python
# Note: the values for the first record is from the initialized state

decision_tree_df.head()
```

```python
# Baseline 
decision_tree_df["weight"].value_counts(normalize=True)
```

```python
# Baseline 
decision_tree_df["height"].value_counts(normalize=True)
```

```python
x = decision_tree_df.drop('animal_type', axis=1)

x
```

```python
y = decision_tree_df.animal_type

# Convert y from float to int
y = y.astype('int')
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 3)
```

### Model Using Entropy

```python
# Set max depth to 3 to make plotting the tree more visible
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3).fit(X_train, y_train)
```

```python
#we can calculate the accuarcy using score
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```

```python
#we can also make a simple confusion matrix
plot_confusion_matrix(clf, X_train, y_train);
```

```python
clf.classes_
```

```python
y_pred = clf.predict(X_test)
```

```python
# Plot the tree
plt.figure(figsize=(10,10))

tree.plot_tree(
    clf, 
    filled=True,
    rounded=True,
    feature_names=["height","weight"], # columns in same order as was trained on
    class_names=[str(x) for x in clf.classes_] # use cl2.classes_ to find class order
);
```

### Model Using Gini

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(x, y, random_state = 3)
```

```python
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf2.score(X_train2, y_train2)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf2.score(X_test2, y_test2)))
```

```python
#we can also make a simple confusion matrix
plot_confusion_matrix(clf2, X_train2, y_train2);
```
